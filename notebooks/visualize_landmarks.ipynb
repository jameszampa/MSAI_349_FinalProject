```python
num_values_dist = {}
for key, value in data.items():
    if len(value) not in num_values_dist:
        num_values_dist[len(value)] = 0
    num_values_dist[len(value)] += 1
```

```python
# pandas df from dict
import pandas as pd
df = pd.DataFrame.from_dict(num_values_dist, orient='index', columns=['count'])
df.plot(kind='bar')
```
